```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
DataTrain=pd.read_csv('train.csv')
DataTest=pd.read_csv('test.csv')
```

```python
#DataTrain.iloc[0]
X_Train=DataTrain.iloc[:,1:20]
Y_Train=DataTrain.iloc[:,20]

X_Test=DataTest.iloc[:,1:20]
#X_Train.head()

X_train, X_val, Y_train, Y_val = train_test_split(
    DataTrain.iloc[:, 1:20], DataTrain.iloc[:, 20], test_size=0.2, random_state=42)
```

```python


# Assuming you have your training data and labels in the following format:
# training_data = numpy array of shape (num_samples, num_features)
# labels = numpy array of shape (num_samples,) containing class labels (0, 1, or 2)

# Calculate the prior probabilities for each class
class_counts = np.bincount(Y_Train)
prior_probabilities = class_counts / len(Y_Train)

# Calculate the means and variances for each class and feature
class_means = []
class_variances = []
for class_label in range(3):
    class_data = X_Train[Y_Train == class_label]
    class_means.append(np.mean(class_data, axis=0))
    class_variances.append(np.var(class_data, axis=0))

# Define the function to calculate the likelihood
def calculate_likelihood(x, mean, variance):
    return (1 / (np.sqrt(2 * np.pi * variance))) * np.exp(-((x - mean) ** 2) / (2 * variance))

# Define the function to make predictions
def predict(test_data):
    predictions = []
    
    for i in range(test_data.shape[0]):
        #print(test_data.iloc[i])
        posterior_probabilities = []
        for class_label in range(3):
            prior = prior_probabilities[class_label]
            likelihood = np.prod(calculate_likelihood(test_data.iloc[i], class_means[class_label], class_variances[class_label]))
            posterior_probabilities.append(prior * likelihood)
        predicted_class = np.argmax(posterior_probabilities)
        predictions.append(predicted_class)
        
    return predictions

# Assuming you have your test data in a numpy array called test_data
classes_pred = predict(X_val)

# test_predictions now contains the predicted class labels for your test data
print(classes_pred)
acc = np.mean(classes_pred == Y_val)*100
print(acc)
```
